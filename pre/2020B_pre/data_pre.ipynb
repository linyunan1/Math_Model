```python
DATA_PATH = "附件三：285号和313号样本原始数据_1.xlsx"
df = pd.read_excel(DATA_PATH)

DATA_PATH = "附件四：354个操作变量信息.xlsx"
df1 = pd.read_excel(DATA_PATH)
```

```python
#最大最小的限幅方法剔除异常值

count = 0

for j in range(df.shape[1]):
    for i in range(df.shape[0]):
        if df.iloc[i,j] > df1.iloc[j,4] or df.iloc[i,j] < df1.iloc[j,3]:
            df.iloc[i,j] = 0
            print("行索引：{i}，列索引：{j}".format(i = i, j = j))
            count += 1
       
count
        
```

```python
#拉依达准则（3σ）准则去除异常值

count = 0

for j in range(df.shape[1]):
    for i in range(df.shape[0]):
        mean = df.iloc[:,j].mean()
        std = df.iloc[:,j].std()
        if abs(df.iloc[i,j] - mean) > 3 * std:
            df.iloc[i,j] = 0
            print("行索引：{i}，列索引：{j}".format(i = i, j = j)) 
            count += 1
            
count
            
```

```python
#样本剔除

COUNT = 0
count_1 = []

for i in range(df.shape[0]):
    count = 0
    for j in range(df.shape[1]):
        if df.iloc[i,j] == 0:
            count += 1
            
    count_1.append(count)    
    
      
for i in range(len(count_1)):
    if count_1[i] > 20:
        df.drop(labels = i, inplace = True)
        COUNT += 1

COUNT
```

```python
#零值处理

count = 0

for i in range(df.shape[0]):
    for j in range(df.shape[1]):
        if df.iloc[i,j] == 0:
            df.iloc[i,j] = (df1.iloc[j,4] + df1.iloc[j,3]) / 2
            count += 1
            
count
```
